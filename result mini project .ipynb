```python
import pandas as pd

df = pd.read_csv('result stude.txt')
print(df.head())
```

```python
df.info()
df.describe()
```

```python
#Calculate Average Score
df['Average'] = df[['maths', 'science', 'english']].mean(axis=1)
df
```

```python
#Find Top Performers
top_students = df.sort_values(by='Average', ascending=False)
print(top_students.head(5))
```

```python
#Pass/Fail Analysis
df['Passed'] = df['Average'] > 40
print(df['Passed'].value_counts())
```

```python
#Save Results
df.to_csv('student_analysis.csv', index=False)
```
